```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df=pd.read_csv("Book1.csv")
x=df.iloc[:,0]
y=df.iloc[:,1]

plt.scatter(x,y)
plt.show()

```

# Gradient Decent

```python
q0=0
q1=0
a=.1
epochs=100

m=float(len(x))

def Cost_Func(q0,q1):
    Z=(1/m)*sum((h-y)**2)
    return Z

for j in range(epochs):
    for i in range(len(x)):
        h=q1*x+q0
        D_q0=(1/m)*sum(h-y)
        D_q1=(1/m)*sum(x*(h-y))
        q0=q0-a*D_q0
        q1=q1-a*D_q1
        print(Cost_Func(q0,q1))

print(q0,q1)
```

```python

Cost_Func(q0,q1)
    
```

```python
h=q0+q1*x
plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(h),max(h)],color='red')
plt.show()
```
